```python
import numpy as np
def split_train_test(data,test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data)*test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices],data.iloc[test_indices]
train_set,test_set = split_train_test(housing,0.2)
print(len(train_set),'train + ',len(test_set),'test')
```

# 独热编码

对于使用LabelEncoder处理过后的非数值特征量，两个临近的特征可能会被学习算法误认为相似。
此时需要使用独热向量对其进行重新编码。
例如:['dog','cat','pig']在使用LabelEncoder后是[0,1,2]。但是机器学习可能认为1和2相似。
但是经过独热编码(OneHot Encoder)后将[0,1,2]编码成[[0001],[0010],[0100]]。


```python
housing_cat_encoded.reshape(-1,1)
#将数组转置 负1行，1列
```

##### np.r_是按列连接两个矩阵，就是把两矩阵上下相加，要求列数相等，类似于pandas中的concat()。
##### np.c_是按行连接两个矩阵，就是把两矩阵左右相加，要求行数相等，类似于pandas中的merge()。

```python
import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.c_[a,b]
print(np.r_[a,b])
print(c)
print(np.c_[c,b])
```

```python
from sklearn.base import BaseEstimator ,TransformerMixin

rooms_ix,bedrooms_ix,population_ix,household_ix=3,4,5,6

class CombinedAttributesAdder(BaseEstimator,TransformerMixin):
    def __init__(self,add_bedrooms_per_room = True):
        self.add_bedrooms_per_room=add_bedrooms_per_room
        #添加默认标志位
    def fit(self,X,y=None):
        return self
        #什么都不做
    def transform(self,X,y=None):
        rooms_per_household = X[:,bedrooms_ix]/X[:,household_ix]
        population_per_household=X[:,population_ix]/X[:,household_ix]
        #计算不同属性的组合
        if self.add_bedrooms_per_room:
        #如果标志位使能
            bedrooms_per_room=X[:,bedrooms_ix]/X[:,rooms_ix]
            return np.c_[X,rooms_per_household,population_per_household,bedrooms_per_room]
            #使用c_按列组合各个属性
        else:
            return np.c_[X,rooms_per_household,population_per_household]
attr_adder = CombinedAttributesAdder(add_bedrooms_per_room = False)
#实例化一个新对象
housing_extra_attribs=attr_adder.transform(housing.values)
#使用DataFrame.values()方法返回nnumpy的类型数据
```

```python
from sklearn.base import BaseEstimator ,TransformerMixin
from sklearn.preprocessing import LabelEncoder

class DataFrameSelector(BaseEstimator ,TransformerMixin):
#创建一个数据帧选择器
    def __init__(self,attribute_name):
        self.attribute_name = attribute_name
    def fit(self,X,y=None):
        return self
    #啥都不干
    def transform(self,X):
        return X[self.attribute_name].values
        #返回attribute对应的帧

class MyLabelBinarizer(TransformerMixin):
    def __init__(self, *args, **kwargs):
        self.encoder = LabelBinarizer(*args, **kwargs)
    def fit(self, x, y=0):
        self.encoder.fit(x)
        return self
    def transform(self, x, y=0):
        return self.encoder.transform(x)

class MyLabelEncoder(BaseEstimator ,TransformerMixin):
    def __init__(self, *args, **kwargs):
        self.encoder = LabelEncoder(*args, **kwargs)
    def fit(self, x, y=0):
        self.encoder.fit(x)
        return self
    def transform(self, x, y=0):
        print('im in')
        a = self.encoder.transform(x)
        print(a)
        return a
```

>这里需要将LabelBinarizer替换成MyLabelBinarizer，原因请看https://blog.csdn.net/qq_36393962/article/details/80038286

```python
from sklearn.pipeline import FeatureUnion

num_attribs = list(housing_num)
cat_attribs = ['ocean_proximity']
#提取出数值数据的列名和文本数据的列名

num_pipeline=Pipeline([
    ('selector',DataFrameSelector(num_attribs)),
    ('imputer',SimpleImputer(strategy='median')),
    ('attribs',CombinedAttributesAdder()),
    ('std_scaler',StandardScaler()),
])
#针对数值数据的转换器流水线

cat_pipeline=Pipeline([
    ('selector',DataFrameSelector(cat_attribs)),
#    ('label_binarizer',MyLabelBinarizer()),
    ('label_encoder',MyLabelEncoder()),
])
#针对文本数据的转换器（独热向量）流水线

full_pipeline=FeatureUnion(transformer_list=[
    ('num_pipeline',num_pipeline),
    ('cat_pipeline',cat_pipeline),
])
#合并为整条流水线

housing_prepared = full_pipeline.fit_transform(housing)
housing_prepared
```

>通过full_pipeline整条流水线，完成对不同类型数据的处理(例如清理空数据，添加更多属性，特征缩放等)，完成机器学习前的数据预处理

```python
from sklearn.linear_model import LinearRegression

line_reg=LinearRegression()
line_reg.fit(housing_prepared,housing_labels)
```

```python
some_data=housing.iloc[:5]
some_labels=housing_labels[:5]
#这里提取housing前5项
print(some_data)
some_data_prepared = full_pipeline.fit_transform(some_data)
print(housing_prepared[1])
print(some_data_prepared[1])
print('Predictions:\t',line_reg.predict(some_data_prepared))
```

>some_data数据集中只有三种ocean_proximity属性，在对some_data进行流水线处理时，ocean_promixity会被独热编码成3个元素。
>而housing_prepared数据集中拥有全部五种ocean_proximity属性，会被独热编码成5个元素。
>这就会导致训练时的数据集(housing_prepared)形状与测试时的数据集(some_data_prepared)形状不一致，导致上面的报错
>解决办法：提取尽可能多的数据以包含全部五种ocean_proximity属性 或者取消独热编码，使用文本映射
